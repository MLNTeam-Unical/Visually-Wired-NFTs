```python
import pickle
import random
```

```python
img_folder = '../input/bigImageDataset/imgs'

id2desc = pickle.load(open("../embeddings/unique2description.pkl","rb"))
id2desc = {k:id2desc[k] for k, v  in id2desc.items() if (v is not None) and (k is not None)}
valid_extensions=["png","jpg"]
id2img = pickle.load(open("../embeddings/unique2path.pkl","rb"))
id2img={k:id2img[k] for k, v  in id2img.items() if (v is not None) and (k is not None) and (v.split(".")[-1] in valid_extensions)}

valid_NFT_ids = list(set(id2desc.keys()) & set(id2img.keys()))


```

```python
desired_ds_size=30000
new_ds_folder="../minDataset"
random.shuffle(valid_NFT_ids)
valid_NFT_ids=valid_NFT_ids[:desired_ds_size]
```

```python
import shutil
import os
from tqdm import tqdm
os.makedirs(f'{new_ds_folder}/imgs')

new_nft2path={}
new_nft2desc={}

for nft in tqdm(valid_NFT_ids):
    shutil.copy(f'{img_folder}/{id2img[nft]}',f'{new_ds_folder}/imgs/{id2img[nft]}')
    new_nft2path[nft]=id2img[nft]
    new_nft2desc[nft]=id2desc[nft]
```

```python
pickle.dump(new_nft2path,open(f'{new_ds_folder}/nft2path.pkl',"wb"))
pickle.dump(new_nft2desc,open(f'{new_ds_folder}/nft2desc.pkl',"wb"))
```
