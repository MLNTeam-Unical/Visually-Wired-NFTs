```python
# Libraries
import pandas as pd
import networkx as nx
```

```python
# Getting data
FOLDER="../graphsV2/NFTLevel-Pruning_V2/(2021, 3)/"
data = pd.read_csv(f"{FOLDER}/edge_list.csv")
data.head()
```

```python
data.shape
```

```python
data.duplicated(keep='first').sum()
```

```python
# Getting percentiles
perc = data['weight'].describe()
perc
```

```python
# Creating graph
threshold = perc['min']
print(f"Creating graph with edge weight >= {threshold}")

data_filtered = data[data.weight >= threshold]
decrease = (data_filtered.shape[0]-data.shape[0])/data.shape[0]
print(f"Original edges: {data.shape[0]} | Remaining edges: {data_filtered.shape[0]} ({decrease:.2%})")

g = nx.from_pandas_edgelist(data_filtered, source='source', target='target', edge_attr='weight', create_using=nx.DiGraph())
print(f"#nodes = {g.number_of_nodes()} | #edges = {g.number_of_edges()}")
```

```python
# Getting out-deg info
out_degrees = [out_deg for node, out_deg in g.out_degree()]
in_degrees = [in_deg for node, in_deg in g.in_degree()]

assert sum(out_degrees) == sum(in_degrees)

avg_out_degree = sum(out_degrees)/len(out_degrees)

print(f"Max in-degree = {max(out_degrees)}")
print(f"Avg out-degree = {avg_out_degree:.2f}")
```
